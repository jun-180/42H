## 문제 2

오전 6시 52분에 집을 떠나서, 가볍게 (km당 8분 15초 소요) 1km를 뛰고, 
이어서 좀 빠르게 (km당 7분 12초 소요) 3km를 뛴 후, 다시 가볍게 1km를 마저 달려 집으로 돌아오면 
집에 도착하는 시간을 계산하여 아래처럼 출력하는 코드를 작성하라.

    집에 도착 시간: 7 시 30 분 6 초

힌트: 정수 나눗셈의 몫과 나머지 연산자 활용

**답**

시, 분, 초를 동시에 다뤄야 하기에 계산을 초 단위로 통일시킨다.
그렇지 않으면 시간의 연산이 복잡해진다.
먼저 출발시간인 6시 52분이 밤 0시를 기준으로 몇 초 지난 시간인지
계산해서 `start` 변수에 할당한다.

````python
start = (6 * 60 + 52) * 60
start
````

아래 두 변수는 1 킬로미터를 가볍게 달릴 때와 좀 빠르게 달릴 때 걸리는 시간(초)을 가리킨다.

````python
slow_running = 8 * 60 + 15 # 가볍게 달릴 때 킬로미터 당 걸리는 시간(초)
fast_running = 7 * 60 + 12 # 좀 빠르게 달릴 때 킬로미터 당 걸리는 시간(초)
````

그러면 달리는 시간을 초로 계산하면 2286초로 나온다.

````python
running_time = (1 + 1) * slow_running + 3 * fast_running
running_time
````

도착시간은 밤 0시를 기준으로 출발시간에 달린 시간을 더하면 된다.

````python
arrival = start + running_time
arrival
````

도착시간을 시, 분, 초로 변환하기 위해 1분은 60초, 1시간은 60분이라는 정보를
이용하여 도착시간에 대해 60으로 나눴을 때의 몫과 나머지 계산을 두 번 연속 적용한다.

- 도착시간의 초 계산: 60으로 나눴을 때의 나머지

````python
arrival_seconds = arrival % 60
arrival_seconds
````

- 도착시간의 분 계산: 60으로 나눴을 때의 몫

````python
arrival_minutes = arrival // 60
arrival_minutes
````

450분은 60분 보다 크기에 시와 분으로 다시 변환해야 한다.
한 시간이 60분이기에 역시 60으로 나눈 몫과 나머지를 계산한다.

- 도착시간의 시 계산: 60으로 나눴을 때의 몫

````python
arrival_hours = arrival_minutes // 60
arrival_hours
````

- 도착시간의 분 계산: 60으로 나눴을 때의 나머지

````python
arrival_minutes = arrival_minutes % 60
arrival_minutes
````

따라서 도착시간은 7시 30분 6초이다. 

````python
print("집에 도착 시간:", arrival_hours, "시", arrival_minutes, "분", arrival_seconds, "초")
````

정리하면 다음과 같다.

````python
start = (6 * 60 + 52) * 60
slow_running = 8 * 60 + 15 # 가볍게 달릴 때 킬로미터 당 걸리는 시간(초)
fast_running = 7 * 60 + 12 # 좀 빠르게 달릴 때 킬로미터 당 걸리는 시간(초)

running_time = (1 + 1) * slow_running + 3 * fast_running
arrival = start + running_time

arrival_seconds = arrival % 60
arrival_minutes = arrival // 60
arrival_hours = arrival_minutes // 60
arrival_minutes = arrival_minutes % 60

print("집에 도착 시간:", arrival_hours, "시", arrival_minutes, "분", arrival_seconds, "초")
````

## 문제 6

대기실에 `N`개의 의자가 일렬로 놓여 있다.
어느 두 사람도 이웃하여 앉지 않을 때, 앉을 수 있는 최대 인원수와
최소 인원수를 출력하는 코드를 구하라.
입력은 100이하의 자연수이고, 최대와 최소는 공백으로 구분하여 출력한다.
단, 자리를 채운 후에는 더 이상 어느 누구도 자리에 앉을 수 없어야 한다.
예를 들어 의자가 8개인 경우 아래 방식으로 앉는 것은 허용되지 않는다.
아래 표에서 `O`은 채워진 자리를, `X`는 빈자리를 가리킨다.

| 불가능 방식 | 이유 |
| :---: | :--- |
| `XXOXOXOX` | 왼쪽 끝에 한 명 더 앉을 수 있음 |
| `XOXXOXXX` | 오른쪽 끝에 한 명 더 앉을 수 있음|

의자가 8 개일 때 의자에 앉을 수 있는 최소 인원과 최대 인원은 다음과 같다.

| 인원 | 앉는 방식 |
| :---: | :---: |
| 최소 3명 | `XOXXOXXO` |
| 최대 4명 | `OXOXOXOX` |

힌트: 2배수와 3배수 여부에 따른 `if ... else ...` 조건문 활용

**답**

`N`개의 의자가 있을 때 예를 들어 맨 왼쪽 의자에 앉은 다음에 한 자리 건널 때마다 앉으면 최대 인원이 앉게 된다.
즉 두 의자당 하나에 앉으면 된다. 만약 `N`이 홀수이면 맨 오른쪽 자리에 한 명이 더 앉을 수 있다.
따라서 `N`이 2배수인지 여부가 중요하다.

- `N`이 2의 배수일 때: 최대 인원은 `N//2`
- `N`이 2의 배수일 때: 최대 인원은 `N//2 + 1`

반면에 최소 인원은 다음과 같이 계산할 수 있다. 
먼저 의자가 1개 또는 2개이면 최대 1명 앉을 수 있다.
이제부터 의자가 3개 이상이라고 가정한다.
먼저 `N = 3`이라고 하자. 그러면 아래 방식으로 1명만 앉을 수 있다.

    XOX

`N > 3`인 경우는 3배수 여부로 결정된다.

- `N`이 3의 배수인 경우: `XOX` 방식을 `N//3` 번 반복한다. 예를 들어 `N = 6`이면 아래 방식으로 최소 인원이 앉는다.

        XOXXOX
        
- `N`이 3의 배수가 아닌 경우: `XOX` 방식을 `N//3` 번 반복한 다음 나머지 하나 또는 두 자리 중에 한 자리에 앉는다.
    예를 들어 `N=8`이면 아래 방식으로 최소 인원이 앉는다.

        XOXXOXXO

위 설명을 정리하여 코드로 작성하면 다음과 같다.

````python
N = int(input("정수를 입력하세요: "))

# 최대 인원
if (N%2 == 0) :
    max = N//2
else :
    max = (N//2) + 1

# 최소 인원
if (N%3 == 0) :
    min = N//3
else :
    min = (N//3) + 1

print("최소:", min, "명")
print("최대:", max, "명")
````

## 문제 7

다음 모양을 출력하는 코드를 작성하라.
두 개의 별 기호 사이에 공백(space)가 사용됨에 주의한다.
공백도 하나의 기호로 취급된다.
단, `print()` 함수와 `while` 반복문은 한 번씩만 활용한다.

```
        * 
      * * 
    * * * 
  * * * * 
* * * * * 
  * * * * 
    * * * 
      * * 
        * 
```

**답**

별 기호와 공백으로 구성된 문자열, 즉 `"* "`이 차례대로 1, 2, 3, 4, 5 번씩,
이후에는 4, 3, 2, 1 번씩 반복된다.
`while` 반복문:  총 9번 출력하며 5번째 출력을 중심으로 처음 4번과 나중 4번이 대칭을 이룬다.

`num`이 1부터 9까지 변할 때 두 개의 공백으로 구성된 문자열 `"  "`와
별과 공백으로 구성된 문자열 `'* '`이 반복되는 횟수를 각각 `spaces` 변수와 `stars` 변수에 할당한 후에
이를 이용하여 출력문을 실행한다.

- `num <= 5` 일 때
    - 공백: 총 `5 - num` 개
    - 별: 총 `num` 개
- `num > 5` 일 때
    - 공백: 총 `num - 5` 개
    - 별: 총 `10 - num` 개

````python
num = 1

while num < 10:
    if num <= 5:
        spaces = 5 - num
        stars = num
    else:
        spaces = num - 5
        stars = 10 - num
        
    print("  " * spaces + "* " * stars)
    
    num += 1
````

## 문제 8

위 문제를 `for` 반복문을 한 번만 활용하여 해결하라.
단, `print()` 함수와 `for` 반복문은 한 번씩만 활용한다.

```
        * 
      * * 
    * * * 
  * * * * 
* * * * * 
  * * * * 
    * * * 
      * * 
        * 
```

힌트: `range()` 함수 활용

**답**

`for` 반복문을 `num` 변수가 1부터 9까지에 대해 실행하도록 한다.
이를 위해 `range(1, 10)` 함수 표현식을 이용한다.

````python
for num in range(1, 10):
    if num <= 5:
        spaces = 5 - num
        stars = num
    else:
        spaces = num - 5
        stars = 10 - num
        
    print("  " * spaces + "* " * stars)
````
